```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

class CodexEntity:
    def __init__(self, name, x, y, z, frequency):
        self.name = name
        # Spatial Coordinates (Rendered Reality)
        self.pos = np.array([x, y, z], dtype=float)
        # Structural Frequency (Topological Reality)
        self.freq = frequency
        self.history = [self.pos.copy()]

def calculate_resonance(freq1, freq2):
    """
    สมการ Topology: คำนวณความเหมือนของ 'รูปแบบปม'
    ผลลัพธ์ (R) คือค่า 0 ถึง 1
    1 = ความถี่ตรงกันเป๊ะ (Same Source)
    0 = คนละความถี่ (No Relation)
    """
    diff = abs(freq1 - freq2)
    # ใช้ Exponential Decay: ยิ่งต่างกันน้อย ค่า R ยิ่งพุ่งสูง
    # R(f) = e^(-k * |f1-f2|)
    k = 2.0  # Sensitivity constant
    resonance = np.exp(-k * diff)
    return resonance

def update_coordinates(entity_source, entity_target, resonance):
    """
    สมการ Pointer Logic: Update พิกัดตามความเข้มข้นของ Resonance
    โดยไม่ใช้แรงโน้มถ่วง (No Gravity Force) แต่ใช้การเขียนทับพิกัด (Overwrite)
    """
    if resonance < 0.01:
        return # ไม่มีการเชื่อมต่อ

    # อัตราการ "วาร์ป" หรือดึงเข้าหากัน (Learning Rate)
    # ยิ่ง Resonance สูง ยิ่งดึงเข้าหากันเร็ว (เหมือน Pointer ชี้ที่เดียวกัน)
    pull_strength = resonance * 0.1

    # Vector การเคลื่อนที่
    direction = entity_target.pos - entity_source.pos

    # Update พิกัด: New_Pos = Old_Pos + (Vector * Strength)
    entity_source.pos += direction * pull_strength

# --- ตั้งค่าการจำลอง (Simulation Setup) ---

# 1. สร้างวัตถุ 2 ชิ้นที่อยู่ไกลกันมากในแกน X, Y, Z
# ตัวอย่าง: Source (จิตเจตจำนง) และ Target (วัตถุแห่งความปรารถนา)
obj_A = CodexEntity("Observer", x=0, y=0, z=0, frequency=500.00)
obj_B = CodexEntity("Manifestation", x=100, y=100, z=100, frequency=500.05)
# หมายเหตุ: ความถี่ต่างกันเพียง 0.05 (Resonance สูงมาก)

steps = 50
print(f"--- เริ่มต้นการจำลอง: {obj_A.name} vs {obj_B.name} ---")
print(f"ระยะห่างเริ่มต้น: {np.linalg.norm(obj_A.pos - obj_B.pos):.2f} หน่วย")

# --- Loop การทำงานของจักรวาล (Main Loop) ---
for t in range(steps):
    # 1. คำนวณ Resonance (ตรวจเช็ค Database)
    R = calculate_resonance(obj_A.freq, obj_B.freq)

    # 2. ปรับพิกัด (Update Pointers)
    # ให้ obj_B ถูกดึงเข้ามาหา obj_A (หรือสลับกัน)
    update_coordinates(obj_B, obj_A, R)

    # เก็บประวัติเพื่อสร้างกราฟ
    obj_B.history.append(obj_B.pos.copy())

    # แสดงผล Log บางช่วง
    if t % 10 == 0 or t == steps - 1:
        dist = np.linalg.norm(obj_A.pos - obj_B.pos)
        print(f"Time {t}: Resonance={R:.4f} | Distance={dist:.2f}")

# --- การแสดงผล (Rendering) ---
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# แปลงข้อมูลประวัติเป็น Array
hist_B = np.array(obj_B.history)

# วาดจุด A (ผู้สังเกต - อยู่กับที่)
ax.scatter(obj_A.pos[0], obj_A.pos[1], obj_A.pos[2], c='red', s=100, label='Source (A)')

# วาดเส้นทางของ B (สิ่งที่ถูกดึงดูด)
ax.plot(hist_B[:,0], hist_B[:,1], hist_B[:,2], c='blue', label='Target Path (B)')
ax.scatter(hist_B[-1,0], hist_B[-1,1], hist_B[-1,2], c='blue', s=100, marker='^', label='Final Pos (B)')

ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')
ax.set_title('The Architect\'s Codex: Frequency Attraction Simulation')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

class CodexEntity:
    def __init__(self, name, x, y, z, base_freq, stability):
        self.name = name
        self.pos = np.array([x, y, z], dtype=float)
        self.base_freq = base_freq
        self.stability = stability # ค่าความนิ่ง (0.0 = มั่วมาก, 1.0 = นิ่งสนิท)
        self.history = [self.pos.copy()]
        self.resonance_log = []

    def get_current_freq(self):
        """
        สร้างความถี่ ณ ขณะนั้น โดยใส่ Noise ตามค่าความเสถียรของจิต
        Noise ยิ่งเยอะ = ความถี่แกว่งไกลจากเป้าหมาย
        """
        noise_factor = (1.0 - self.stability) * 5.0 # ปรับความรุนแรงของ Noise
        jitter = np.random.normal(0, noise_factor)
        return self.base_freq + jitter

def calculate_resonance_dynamic(freq1, freq2):
    # ยิ่งค่า K สูง ระบบยิ่งเข้มงวด (ต้องตรงกันเป๊ะถึงจะดึงดูด)
    k = 0.5
    diff = abs(freq1 - freq2)
    return np.exp(-k * diff)

def update_simulation(source, target, steps=100):

    target_freq = source.base_freq # เป้าหมายคือต้องจูนให้ตรง Source

    for t in range(steps):
        # 1. รับค่าความถี่ที่มี Noise ปน
        current_freq_source = source.get_current_freq()

        # 2. คำนวณ Resonance (ความแรงในการดึงดูด ณ วินาทีนั้น)
        R = calculate_resonance_dynamic(current_freq_source, target_freq)
        source.resonance_log.append(R)

        # 3. Update Position (ถ้า R น้อย แรงดึงจะอ่อนมาก จนแทบไม่ขยับ)
        if R > 0.05: # Threshold: ถ้าใจลอยเกินไป การดึงดูดจะไม่เกิดเลย
            pull_strength = R * 0.05 # Speed factor
            direction = target.pos - source.pos
            source.pos += direction * pull_strength

        source.history.append(source.pos.copy())

# --- Setup Comparison ---

# เป้าหมายปลายทาง (Goal) อยู่ที่พิกัด (100, 100, 100)
Goal_Target = CodexEntity("The Desire", 100, 100, 100, base_freq=500.0, stability=1.0)

# ผู้ทดลองที่ 1: จิตนิ่ง (Stability 98%)
Master = CodexEntity("Focused Mind", 0, 0, 0, base_freq=500.0, stability=0.98)

# ผู้ทดลองที่ 2: จิตฟุ้งซ่าน (Stability 60%) - มีความสงสัย/กลัวปนอยู่
Novice = CodexEntity("Distracted Mind", 0, 0, 0, base_freq=500.0, stability=0.60)

# Run Simulation
steps = 200
update_simulation(Master, Goal_Target, steps)
update_simulation(Novice, Goal_Target, steps)

# --- Visualization ---
fig = plt.figure(figsize=(12, 6))

# Plot 1: 3D Trajectory (เส้นทางการดึงดูด)
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
hist_M = np.array(Master.history)
hist_N = np.array(Novice.history)

ax1.plot(hist_M[:,0], hist_M[:,1], hist_M[:,2], c='green', linewidth=2, label='Focused Mind (Stable)')
ax1.plot(hist_N[:,0], hist_N[:,1], hist_N[:,2], c='red', linestyle='--', linewidth=1, label='Distracted Mind (Noisy)')
ax1.scatter(100, 100, 100, c='blue', s=100, marker='*', label='Target Reality')
ax1.set_title('Path of Attraction (x,y,z)')
ax1.legend()

# Plot 2: Resonance Stability (ความเสถียรของการเชื่อมต่อ)
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(Master.resonance_log, c='green', alpha=0.6, label='Focused Resonance')
ax2.plot(Novice.resonance_log, c='red', alpha=0.6, label='Distracted Resonance')
ax2.set_title('Resonance Signal Strength over Time')
ax2.set_xlabel('Time Steps')
ax2.set_ylabel('Connection Strength (0-1)')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Observer:
    def __init__(self, name, target_pos, willpower, stability):
        self.name = name
        self.target_pos = target_pos # ตำแหน่งที่อยากให้วัตถุไป
        self.willpower = willpower   # ความแรงของความปรารถนา (Amplitude)
        self.stability = stability   # ความนิ่ง (Consistency) 0.0 - 1.0

    def exert_influence(self):
        """
        ส่งค่า 'แรงดึง' ออกไป
        Noise จะลดทอนพลังของ Willpower ลงชั่วคราว
        """
        # สร้าง Noise ตามความไม่เสถียร
        noise = np.random.normal(0, (1.0 - self.stability) * 2.0)

        # พลังที่ส่งออกไปจริง = Willpower + Noise (อาจจะเสริมหรือหักล้างก็ได้)
        effective_power = self.willpower + noise

        # พลังต้องไม่ติดลบ (อย่างน้อยต้องเป็น 0)
        return max(0, effective_power)

# --- Simulation Setup ---

# วัตถุเริ่มต้นที่ตำแหน่ง 0
object_position = 0
history = [object_position]

# Observer A: แรงเยอะ แต่จิตใจวอกแวก (Power=10, Stable=0.4)
# อยากให้กราฟพุ่งขึ้นไปที่ +100
Player_A = Observer("Mr. Passion", target_pos=100, willpower=12.0, stability=0.4)

# Observer B: แรงน้อย แต่จิตใจนิ่งสงบ (Power=6, Stable=0.9)
# อยากให้กราฟดิ่งลงไปที่ -100
Player_B = Observer("Mr. Discipline", target_pos=-100, willpower=7.0, stability=0.95)

steps = 200

print(f"--- เริ่มการปะทะ: {Player_A.name} vs {Player_B.name} ---")

# --- Loop ---
for t in range(steps):
    # 1. รับค่าพลังจากทั้งสองฝ่าย
    pull_A = Player_A.exert_influence()
    pull_B = Player_B.exert_influence()

    # 2. คำนวณ Vector (ทิศทาง * พลัง)
    # A ดึงไปทางบวก, B ดึงไปทางลบ
    # เราใช้ Factor 0.05 เพื่อให้การขยับค่อยเป็นค่อยไป (Inertia)
    move_A = (Player_A.target_pos - object_position) * (pull_A * 0.005)
    move_B = (Player_B.target_pos - object_position) * (pull_B * 0.005)

    # 3. ผลลัพธ์สุทธิ (Net Movement)
    delta = move_A + move_B
    object_position += delta

    history.append(object_position)

# --- Visualization ---
plt.figure(figsize=(10, 6))

# วาดเส้นแบ่งเขตแดน
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5, label='Neutral Zone')
plt.axhline(y=100, color='green', linestyle=':', alpha=0.5, label='Goal A')
plt.axhline(y=-100, color='red', linestyle=':', alpha=0.5, label='Goal B')

# วาดเส้นกราฟความจริง (Reality Line)
plt.plot(history, color='purple', linewidth=2, label='Manifested Reality')

# ตกแต่งกราฟ
plt.title(f'Consensus Reality: {Player_A.name} (High Power/Low Stable) vs {Player_B.name} (Low Power/High Stable)')
plt.xlabel('Time Steps')
plt.ylabel('Reality State (Position)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

```python
import time
import math

class RealityCluster:
    def __init__(self, name, mass_load):
        self.name = name
        self.mass_load = mass_load # Load 0.0 to 1.0 (1.0 = Black Hole)
        self.local_time = 0.0

    def process_frame(self, server_tick_duration):
        """
        จำลองการประมวลผล 1 เฟรมของความเป็นจริง
        """
        # 1. คำนวณ Overhead จากการ Render มวล
        # ยิ่งมวลเยอะ ยิ่งต้องเสียเวลา Loop เพื่อประมวลผลข้อมูล (Processing Cost)
        processing_cost = 0
        if self.mass_load > 0:
            # จำลองความหน่วง (Simulated CPU Load)
            # ในความเป็นจริง คือการเสีย Clock Cycles ไปกับ Gravity Calculation
            factor = 1 / math.sqrt(1 - self.mass_load)
            processing_cost = (factor - 1) * server_tick_duration

        # 2. Update เวลา
        # เวลาที่ผ่านไปจริงในมุมมองของ Cluster นี้ จะช้าลงเมื่อเทียบกับ Server
        # เพราะเสียเวลาไปกับ Processing Cost

        # สมมติ Server เวลาผ่านไป 1 วินาที
        # แต่ Cluster นี้รู้สึกว่าผ่านไปแค่...
        time_dilation_factor = math.sqrt(1 - self.mass_load)
        self.local_time += server_tick_duration * time_dilation_factor

        return self.local_time

# --- Simulation Setup ---

# โซน A: อวกาศว่างเปล่า (Load 0%)
Space_Station = RealityCluster("Deep Space Station", mass_load=0.0)

# โซน B: ดาวเคราะห์ยักษ์ (Load 50% - แรงโน้มถ่วงสูงมาก)
Miller_Planet = RealityCluster("Super-Gravity Planet", mass_load=0.5)

# โซน C: ขอบหลุมดำ (Load 99% - เครื่องเกือบค้าง)
Event_Horizon = RealityCluster("Black Hole Edge", mass_load=0.99)

server_time = 0
duration = 10 # รัน Server ไป 10 วินาที

print(f"{'Server Time':<12} | {'Space Station':<15} | {'Miller Planet':<15} | {'Black Hole':<15}")
print("-" * 65)

for i in range(duration + 1):
    t_space = Space_Station.process_frame(1.0)
    t_planet = Miller_Planet.process_frame(1.0)
    t_blackhole = Event_Horizon.process_frame(1.0)

    print(f"{i:<12} | {t_space:<15.2f} | {t_planet:<15.2f} | {t_blackhole:<15.2f}")

    # หมายเหตุ: ในโค้ดจริงเราไม่ได้หน่วงเวลา Python (sleep)
    # แต่เราคำนวณ "Local Time Age" ที่สะสมได้

print("-" * 65)
print("สรุปความต่างของเวลา (Time Slippage):")
print(f"บนอวกาศผ่านไป: {Space_Station.local_time:.2f} วินาที")
print(f"บนดาวเคราะห์ผ่านไป: {Miller_Planet.local_time:.2f} วินาที (หายไป {10 - Miller_Planet.local_time:.2f} วิ)")
print(f"ขอบหลุมดำผ่านไป: {Event_Horizon.local_time:.2f} วินาที (หายไป {10 - Event_Horizon.local_time:.2f} วิ)")
```

```python
import random
import time

class Soul:
    def __init__(self, soul_id):
        self.id = soul_id
        self.karma_score = 0     # สะสมบุญ/บาป (Log History)
        self.experience = 0      # เลเวลทางจิตวิญญาณ
        self.memory_cache = []   # ความจำชาติปัจจุบัน (จะถูกลบ)

    def sync_to_cloud(self):
        """
        กระบวนการหลังตาย: Upload Karma, ล้าง Memory
        """
        print(f"\n[SYSTEM]: Uploading Soul ID {self.id} to Cloud...")
        time.sleep(1)

        # Karma กำหนดโชคชะตาชาติหน้า
        final_score = self.karma_score

        # Garbage Collection (ล้างความจำ)
        print(f"[SYSTEM]: Clearing Local Cache (Deleting {len(self.memory_cache)} memories)...")
        self.memory_cache = []

        return final_score

class AvatarBody:
    def __init__(self, soul, environment_difficulty):
        self.soul = soul
        self.health = 100
        self.environment = environment_difficulty # ความยากง่ายของชีวิต
        print(f"--- New Life Started: Environment Level {self.environment}/10 ---")

    def live_life(self):
        """
        จำลองการใช้ชีวิต: สร้างกรรม (Action)
        """
        actions = random.randint(-10, 20) # ทำดีบ้าง ชั่วบ้าง
        self.soul.karma_score += actions
        self.soul.memory_cache.append("Some random memory")
        self.health = 0 # จบชีวิต
        print(f"Event: Life lived. Final Karma: {self.soul.karma_score}")

def respawn_system(soul):
    # 1. ตายและ Sync ข้อมูล
    karma = soul.sync_to_cloud()

    # 2. คำนวณสถานที่เกิดใหม่ (Algorithm)
    # Karma สูง -> Difficulty ต่ำ (ชีวิตสบาย)
    # Karma ต่ำ -> Difficulty สูง (ชีวิตลำบาก)

    if karma > 50:
        new_difficulty = 1  # Easy Mode
        realm = "Heavenly/Prosperous Realm"
    elif karma > 0:
        new_difficulty = 5  # Normal Mode
        realm = "Human Realm"
    else:
        new_difficulty = 9  # Hard Mode
        realm = "Lower Realm"

    print(f"[SYSTEM]: Matching Frequency... Target Realm: {realm}")

    # 3. สร้างร่างใหม่ (New Instance)
    return AvatarBody(soul, new_difficulty)

# --- Simulation ---

# 1. กำเนิดจิตวิญญาณ (First Instance)
My_Soul = Soul(soul_id="User_777")

# 2. ชาติที่ 1
Current_Body = AvatarBody(My_Soul, environment_difficulty=5)
Current_Body.live_life() # ใช้ชีวิตจนตาย

# 3. การเวียนว่ายตายเกิด (Loop)
# จำลองการเกิดใหม่ 3 ชาติ
for life in range(2, 5):
    Current_Body = respawn_system(My_Soul)
    Current_Body.live_life()
    print("------------------------------------------------")
```
